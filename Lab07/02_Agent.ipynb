**Andrew Ng**: An agentic AI workflow is an iterative, **multi-step process** where AI agents autonomously plan, reason, and execute tasks with minimal human oversight, moving beyond a single-prompt output to achieve a better final result. It involves breaking down a complex goal into smaller, more manageable steps, allowing the AI to use its **reasoning** and **tools**, **critique** its own work, and **revise** its output until a **high-quality** outcome is achieved.

### Example 1: Customer-service
Query: What is your refund policy?

Reponse: (via RAG): Refund are available within 30 days of purchase. 


Query: Can I get a refund for my order?
Response (via AI agent):
- The agent retrieves refund policy using RAG.
- The agents asks: "Can you provide your order number?"
- It queries an API to check the order details.
- It confirms: "Your order qualifies for a refund. The amount will be processed in 3-5 business days."

### Example 2: Essay-writing
None agentic:
Please type out an essay on topic X from start to finish in one go, without using backspace.

Agentic workflow:
Write an essay outline on topic X.
Do you need any web research?
Write a first draft.
Consider what parts need revision or more research. 
Revise your draft.
...

Revise <--> Thinking/Research
