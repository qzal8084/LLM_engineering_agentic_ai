# Generate Image
- Connect to AvalAI
- Use its API to generate an image from a text


```python
import os
from openai import OpenAI
from dotenv import load_dotenv
```

```python
load_dotenv()
aval_api_key=os.getenv("AVALAI_API_KEY")
```

```python
client = OpenAI(
    api_key=aval_api_key,
    base_url="https://api.avalai.ir/v1",
)

response = client.images.generate(
    model="dall-e-3",
    prompt="یک خانم کوهنورد در حال صعود به قله دماوند در یک روز پاییزی",
    n=1,
    size="1024x1024",
    response_format="url",
)

image_url = response.data[0].url
print(image_url)
```
