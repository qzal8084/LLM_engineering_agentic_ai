# Part 2: Generators

We use Generators in this course because AI models can stream their results â€” and generators let us handle that kind of behavior nicely.

In Python, generators are a clean way to build iterators â€” objects you can loop through one element at a time. Theyâ€™re great when you want to process data piece by piece, instead of loading everything into memory all at once.

Key Ideas

Generators let you loop over data without creating a full list in memory. Instead, they produce one value at a time â€” only when you need it.

The magic happens with the yield keyword. When a function uses yield instead of return, it becomes a generator function. Each time yield runs, it pauses the function, saves its state, and hands back a value.

This approach is called lazy evaluation â€” values are only computed when requested, which keeps things memory-efficient.

In short, generators are a simple but powerful way to handle large or even infinite data streams efficiently and elegantly in Python.

````python
# Simple Function
def number_return():
    return_list = []
    for i in range(1, 6):
        return_list.append(i)
    return return_list
````

````python
# Simple Generator Function
def number_generator():
    for i in range(1, 6):
        yield i
````

````python
# Create a generator
gen = number_generator()

# Use a loop to get values from the generator
for number in gen:
    print(number)
````

````python
def infinite_generator():
    i = 1
    while True:
        yield i
        i += 1
````

````python
# Using the generator
gen = infinite_generator()
print(next(gen))  
print(next(gen))  
# Keep calling `next(gen)` to get the next number in the sequence.
````

````python
# First define a generator; it looks like a function, but it has yield instead of return

import time

def come_up_with_fruit_names():
    for fruit in fruits:
        time.sleep(1) # thinking of a fruit
        yield fruit
````

````python
# Then use it

for fruit in come_up_with_fruit_names():
    print(fruit)
````

````python
# Here's another one

def authors_generator():
    for book in books:
        if book.get("author"):
            yield book.get("author")
````

````python
# Use it

for author in authors_generator():
    print(author)
````

````python
# Here's the same thing written with list comprehension

def authors_generator():
    for author in [book.get("author") for book in books if book.get("author")]:
        yield author
````

````python
# Use it

for author in authors_generator():
    print(author)
````

````python
# Here's a nice shortcut
# You can use "yield from" to yield each item of an iterable

def authors_generator():
    yield from [book.get("author") for book in books if book.get("author")]
````

````python
# Use it

for author in authors_generator():
    print(author)
````

````python
# And finally - we can replace the list comprehension with a set comprehension

def unique_authors_generator():
    yield from {book.get("author") for book in books if book.get("author")}
````

````python
# Use it

for author in unique_authors_generator():
    print(author)
````

````python
# And for a bit of fun â€” hit the stop button in the toolbar when youâ€™ve had enough!
# Itâ€™s kind of like we built our own mini language model... just not a very big one.
# Notice how it prints one letter at a time instead of a full word?
# If youâ€™re not sure why, try taking out the word â€œfromâ€ everywhere in the code and see what happens.

import random
import time

pronouns = ["I", "You", "We", "They"]
verbs = ["like", "hate", "dance with", "talk about", "lose", "collect", "hide", "chase", "admire", "borrow", "argue with", "draw", "cook", "sing to", "dream about"]
adjectives = ["silly", "shiny", "grumpy", "bouncy", "sleepy", "colorful", "weird", "tiny", "loud", "fluffy", "sticky", "happy", "fuzzy", "messy", "sparkly"]
nouns = ["bananas", "robots", "penguins", "cupcakes", "dragons", "socks", "toasters", "aliens", "umbrellas", "tacos", "marshmallows", "pencils", "donuts", "clouds"]

def infinite_random_sentences():
    while True:
        yield from random.choice(pronouns)
        yield " "
        yield from random.choice(verbs)
        yield " "
        yield from random.choice(adjectives)
        yield " "
        yield from random.choice(nouns)
        yield ". "

for letter in infinite_random_sentences():
    print(letter, end="", flush=True)
    time.sleep(0.02)
````

# Part 3: Decorators
A decorator in Python is a special kind of function that modifies the behavior of another function (or method, or class) â€” without permanently changing its code.

Think of it as wrapping another function to add something new before or after it runs.


````python
def greet():
    "A simple function"
    print("Hello!")

print(greet.__name__) 
print(greet.__doc__) 
````

````python
def my_decorator(func):
    def wrapper():
        print("Something before the function runs")
        func()
        print("Something after the function runs")
    return wrapper
````

````python
@my_decorator
def greet_decorated():
    print("Hello!")
````

````python
greet_decorated()
````

````python
print(greet_decorated.__name__) 
print(greet_decorated.__doc__) 
````

````python
greet_decorated = my_decorator(greet)
````

````python
greet_decorated()
````

````python
from functools import wraps
def my_decorator(func):
    @wraps(func)
    def wrapper():
        print("Something before the function runs")
        func()
        print("Something after the function runs")
    return wrapper

@my_decorator
def greet_decorated_fixed():
    print("Hello again!")

print(greet_decorated_fixed.__name__) 
print(greet_decorated_fixed.__doc__) 
````

2. Class Methods

Class methods are methods that belong to the class itself, rather than any instance of the class. They are defined using the @classmethod decorator and take cls (representing the class) as their first parameter instead of self (representing the instance).

````python
from typing import List

class Book:
    title: str
    author: str

    def __init__(self, title: str, author: str):
        self.title = title
        self.author = author

    def get_description(self) -> str:
        return f"'{self.title}' by {self.author}"

    @classmethod
    def from_string(cls, info: str) -> Book:
        title, author = info.split(', ')
        return cls(title, author)
    
    @classmethod
    def create_multiple(cls, titles_and_authors: List[tuple]) -> List['Book']:
        """Class method that returns a list of Book instances from a list of title-author pairs."""
        return [cls(title, author) for title, author in titles_and_authors]
````

````python
# Using the class method and the subclass
book = Book.from_string("ÙÛŒÙ‡ Ù…Ø§ ÙÛŒÙ‡, Ù…ÙˆÙ„Ø§Ù†Ø§")
print(book.title)
print(book.author)
````

3. Static method

A @staticmethod does not receive an implicit first argument (neither self nor cls). It's a method that doesn't access or modify class or instance state.
Here, add is a utility function that logically belongs to the class 
but doesn't interact with class or instance data.

````python
class MathOperations:
    @staticmethod
    def add(x, y):
        return x + y


result = MathOperations.add(5, 3)
print(result) 
````

4. Subclasses

Subclasses allow you to create a class that inherits attributes and methods from another class. This is a form of polymorphism, where the subclass can either inherit or override behavior from the parent (or "superclass"). Itâ€™s useful when you want to build on existing functionality.

````python
class LibraryItem:
    title: str
    year: int

    def __init__(self, title: str, year: int):
        self.title = title
        self.year = year

    def get_info(self) -> str:
        return f"{self.title} ({self.year})"

class Book(LibraryItem):
    author: str

    def __init__(self, title: str, year: int, author: str):
        super().__init__(title, year)
        self.author = author

    def get_info(self) -> str:
        return f"'{self.title}' by {self.author} ({self.year})"

    @classmethod
    def from_string(cls, info: str) -> 'Book':
        title, author, year = info.split(', ')
        return cls(title, int(year), author)

# Using the class method and the subclass
book = Book.from_string("The Catcher in the Rye, J.D. Salinger, 1951")
print(book.get_info())
````

5. Pydantic

Pydantic is a data validation and settings management library using Python type annotations.

Itâ€™s built around the concept of defining schemas as Python classes. These schemas ensure the data matches the expected types, and optionally transforms or validates the data automatically.

Think of it as:
ðŸ”Ž "I define what a piece of data should look like â€” Pydantic makes sure it does."

````python
from pydantic import BaseModel

class Book(BaseModel):
    title: str
    author: str
    pages: int
    published: bool = True  # Default value

book = Book(title="1984", author="George Orwell", pages=328)
print(book)
````

````python
book = Book(title="Dune", author="Frank Herbert", pages="412")
print(book.pages)  # int: 412, even though input was a string
````

````python
Book(title="Dune", author="Frank Herbert", pages="not_a_number")
# raises ValidationError!
````

````python
fields = {
    "title": "The Hobbit",
    "author": "J.R.R. Tolkien",
    "pages": 310
}

book = Book(**fields)
print(book)
````

````python
print(book.model_dump())
````
